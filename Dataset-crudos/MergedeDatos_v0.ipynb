#Data frame principal
Cuenta con 42 estaciones monitoreadas 4 veces en el año 2022.

```python
monitoreo2022_df = pd.read_csv('https://raw.githubusercontent.com/MaricelSantos/Mentoria-Diplodatos-2024/main/Dataset-crudos/agc_y_riodelaplata2022_final_2023.csv', sep=";")
```

## Columnas Data Frame principal
* Las columnas 'orden', 'sitio' y 'codigo' se relacionan con las estaciones de monitoreo. El monitoreo abarca de Tigre a Berisso, porque lo que el orden 1 corresponde a estaciones de Tigre mientras que 42 a Berisso. A su vez la columna sitio informa exactamente de que lugar se trata mientra que la última la códifica.
* campaña: estación del año en que fue monitoreada.

* tem_agua: temperatura del agua [°C]
* tem_aire: temperatura del aire [°C]
* od: óxigeno disuelto [mg/L]
* pH: potencial hidrogeno escala 1-14
* olores: presencia/ausencia
* color: presencia/ausencia
* espumas: presencia/ausencia
* mat_susp: material suspendido - presencia/ausencia
* colif_fecales_ufc_100ml: coliformes fecales [U.F.C/100mL] (unidades formadoras de colonia/100mL)
* escher_coli_ufc_100ml: Escherichia coli [U.F.C/100mL]
* enteroc_ufc_100ml: Enterococos [U.F.C/100mL]
* nitrato_mg_l: Nitratos (NO3-) [mg/L]
* nh4_mg_l: Amonio (NH4+) [mg/L]
* p_total_l_mg_l: Fósforo Total [mg/L]
* fosf_ortofos_mg_l: Fosfatos (PO4) [mg/L]
* dbo_mg_l: demanda biológica de óxigeno [mg/L]
* dqo_mg_l: demanda química de óxigeno [mg/L]
* turbiedad_ntu: Turbidez NTU: Unidades Nefelométricas de Turbidez
* hidr_deriv_petr_ug_l: Hidrocarburos derivados del Petróleo [ug/L]
* cr_total_mg_l: Cromo Total [mg/L]
* cd_total_mg_l: Cadmio Total [mg/L]
* clorofila_a_ug_l: Clorofila ‘a’ [ug/L]
* microcistina_ug_l: Microcistina  [ug/L]
* ica: Indice de calidad de agua. Deberiva de ecuaciones que relacionan las características organolepticas, biologicas, químicas y físicas del agua.
* calidad_de_agua: Clasificación que depende del ICA

# Sustancias Químicas en Superficie

```python
sustquimicas_df = pd.read_csv('https://raw.githubusercontent.com/MaricelSantos/Mentoria-Diplodatos-2024/main/Dataset-crudos/agc_l_monit_sust_qcas_2021_2022.csv', sep=";")
sustquimicas22_df = sustquimicas_df[sustquimicas_df['año'] == 2022]
sustquimicas22_df = sustquimicas22_df.reset_index(drop=True)
sustquimicas22drop = sustquimicas22_df.drop(columns=['año', 'campaña','clasificación', 'ld', 'lc', 'unidad']) #Sólo está tomado en primavera
```

```python
sustquimicas22drop
```

```python
df_transpuesto = sustquimicas22drop.transpose()
df_transpuesto.reset_index(inplace=True)
df_transpuesto.columns = df_transpuesto.iloc[0]
df_transpuesto = df_transpuesto.rename(columns={'parámetro': 'codigo', 'a_BHC': 'a_BHC_x',
                                                'b_BHC': 'b_BHC_x', 'g_BHC': 'g_BHC_x', 'Clordano': 'Alfa_clordano' ,
                                                'Carvedilol' : 'Carvedilol_x', 'Propinox': 'Propinox_x', 'Salbutamol':'Salbutamol_x',
                                                'pp_DDD': 'pp_DDD_x', 'op_DDE':'op_DDE_x', 'pp_DDE': 'pp_DDE_x', 'op_DDT': 'op_DDT_x',
                                                'pp_DDT':'pp_DDT_x', 'Endosulfanes':'Endosulfanes_x', 'Endrina': 'Endrina_x',
                                                'Epóxido_de_heptacloro_(isóm_B)': 'Epóxido_de_heptacloro_(isóm_B)_x'})
df_transpuesto= df_transpuesto.drop(0)
df_transpuesto['campaña'] = 'Primavera'
renombrar1 = {'Dureza': 'Dureza_x', 'Fluoruros': 'Floruros_x', 'Sólidos_totales': 'Sólidos_totales_x'}
df_transpuesto = df_transpuesto.rename(columns=renombrar1 )
df_transpuesto
```

```python
merged2_data = pd.merge(merged1_data, df_transpuesto, left_on=['codigo','campaña'], right_on=[df_transpuesto['codigo'].str.upper().str.replace('_', ' '), 'campaña'], how= 'outer') # how='left'
merge2= merged2_data.drop(columns=['codigo_x', 'codigo_y','sitio'])
merge2
```

#Sustancias Químicas en Sedimentos

```python
sedimentos_df = pd.read_csv('https://raw.githubusercontent.com/MaricelSantos/Mentoria-Diplodatos-2024/main/Dataset-crudos/agc_i_monit_sust_qcas_sedim_invie2022_ver2023.csv', sep=";")
sedimentos22_df = sedimentos_df[sedimentos_df['año'] == 2022]
sedimentos22_df.loc[sedimentos22_df['campaña_mes'] == 'Agosto', 'campaña_mes'] = 'Invierno'
sedimentos22_df = sedimentos22_df.reset_index(drop=True)
sedimentos22drop = sedimentos22_df.drop(columns=['año', 'campaña_mes','clasificación', 'ld', 'lc', 'unidades']) #Sólo está tomado en invierno
```

```python
df_sedimentosT = sedimentos22drop.transpose()
df_sedimentosT.reset_index(inplace=True)
df_sedimentosT.columns = df_sedimentosT.iloc[0]
renombrar = {'parámetro': 'codigo', 'Sulfato': 'Sulfatos', 'Alacloro': 'Alacloro_y',
             'Diuron': 'Diuron_y', 'Fenitrotión': 'Fenitrotion_y', 'Linuron': 'Linurón',
             'Mancozeb': 'Mancozeb_y', 'Tiacloprid': 'Tiacloprid_y', 'Tiametoxam': 'Tiametoxam_y',
             'Trifluralina': 'Trifluralina_y', 'Metil_triclosan' : 'Metiltriclosan', 'Emamectina': 'Emamectina_y',
             'Enroflaxacina':'Enroflaxacina_y', 'Metformina': 'Metformina_y', 'Etinil_estradiol':'Etinilestradiol',
             'HAPs_TOTALES': 'HAPs_totales', '4-Cloro-3-metilfenol': '4-Cloro-3-metilfenol_y', 'Estradiol\xa0': 'Estradiol',
             'Heptacloro_epóxido': 'Epóxido_de_heptacloro_(isóm_A)', 'Alfa_endosulfan' : 'Alfa_endosulfan_y', 'DDE' : 'DDE_y',
             'Endrin' : 'Endrin_y', 'Beta_endosulfan': 'Beta_endosulfan_y', 'DDD': 'DDD_y', 'Endosulfán_sulfato': 'Endosulfán_sulfato_y',
             'DDT': 'DDT_y', 'Endrin_cetona': 'Endrin_cetona_y', 'Cromo': 'Cromo_y' , 'Manganeso' : 'Manganeso_y', 'Hierro' : 'Hierro_y',
             'Plata' : 'Plata_y', 'Cadmio' : 'Cadmio_y', 'Uranio': 'Uranio_y', 'Carbono_orgánico_total': 'Carbono_orgánico_total_y',
             'Nitrógeno_total' : 'Nitrógeno_total_y', 'Cianuro': 'Cianuro_y', 'SAAM':'SAAM_y', 'Fósforo_total': 'Fósforo_total_y',
             'Amonio':'Amonio_y', 'Nitratos':'Nitratos_y', 'Fenoles_totales':'Fenoles_totales', 'Alfa_HCH':'Alfa_HCH_y', 'Beta_HCH':'Beta_HCH_y',
             'Gama_HCH':'Gama_HCH_y'}
df_sedimentosT = df_sedimentosT.rename(columns=renombrar )
palabra_a_eliminar = 'Congener'
columnas_a_eliminar = [col for col in df_sedimentosT.columns if col.startswith(palabra_a_eliminar)]
df_sedimentosT = df_sedimentosT.drop(columns=columnas_a_eliminar)
df_sedimentosT= df_sedimentosT.drop(0)
df_sedimentosT['campaña'] = 'Invierno'
df_sedimentosT
```

```python
merged3_data = pd.merge(merge2, df_sedimentosT, left_on=['codigo','campaña'], right_on=[df_sedimentosT['codigo'].str.upper().str.replace('_', ' '), 'campaña'], how= 'outer') # how='left'
merge3= merged3_data.drop(columns=['codigo_x', 'codigo_y'])
merge3
```
