# Concatenación
Sólo utilizo 24,23, 22 y 21 porque son los más completos. Pierdo información o agrego muchos faltantes si utilizo los otros. Además la información de población corresponde a senso 2022. No sería correcto irnos tan atrás.

```python
_main_df = pd.concat([data_21, data_22, data_23, data_24], join= "inner")
_main_df.shape
_main_df["año"].unique()
```

```python
# Función para convertir valores
def convert_to_float(val):
    try:
        return float(val)
    except ValueError:
        return np.nan

#Hacer una copia del df para no perder la información
_main_copy = _main_df.copy()
_main_copy.drop(columns='fecha',inplace=True)
# Aplicar la función a todas las columnas del DataFrame
for _col in _main_copy.columns:
    if _col != 'fecha':
        _main_copy[_col] = _main_copy[_col].apply(convert_to_float)


#Quiero identificar las columnas que son 100% NaN

_columns_only_nan_main = []

for _col in _main_copy.columns:
    if _main_copy[_col].isna().all():
        _columns_only_nan_main.append(_col)

print("Columnas que son solo NaN:")
print(_columns_only_nan_main)


# Eliminar columnas que contienen solo valores NaN
_main_copy.dropna(axis=1, how='all', inplace=True)
```

```python
# Calcular la proporción de NaN por variable en las columnas que son solo NaN
total_observations = len(_main_copy)
nan_proportions = {_col: _main_copy[_col].isna().sum() / total_observations for _col in _main_copy.columns}

# Redondea los valores almacenados en nan_proportions
nan_proportions_rounded_main = {_col: round(v, 3) for _col, v in nan_proportions.items()}


print("\nProporción de NaN por variable:")
print(nan_proportions_rounded_main)
```

```python
print('Veamos la Missing Matrix para ver la densidad de nuestros datos')
plt.figure(figsize=(6, 3))
msno.matrix(_main_copy,color=[0,0,0.5])
```

```python
print('Tambien podemos ver una visualizacion simple de la nulidad por columna con el siguiente grafico.\nEn este grafico no se puede individualizar cada observacion, ergo no podemos ver la continuidad de cada fila.')

msno.bar(_main_copy)
```
